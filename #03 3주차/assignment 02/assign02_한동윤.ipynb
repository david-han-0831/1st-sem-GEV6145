```python
# 1. [10] 수강생의 학번(ID), 성명(name), 성적(점수:score), 전공(major), 과목(couse)을 포함한
#    필요한 속성들을 저장할 수 있는 class를(Student) 정의하고,

# 2. [10] 평균성적을 출력하는 method 및 추가로 필요한 method 혹은 function을 정의합니다.
#    모든 학생은 GEV6145를 수강하고 있는 것으로 가정합니다.



# Excel 파일을 읽기위한 pandas import
import pandas as pd

# 80점 이상 100점 이하 사이의 random number를 생성하기 위한 random import
import random

# class 생성
class Student:

    # 초기화 함수
    # 학번(ID), 성명(name), 성적(점수:score), 전공(major), 과목(course) 지정
    def __init__(self, ID, name, score, major, course):
        self.ID = ID
        self.name = name
        self.score = score
        self.major = major
        self.course = course

    # ["name"] 으로 접근하기 위한 method
    # 9. 문제에서 사용
    def __getitem__(self, key):
        if key == "name":
            print(f"{self.ID}, {self.score}")

    # 모든 학생은 GEV6145를 수강하고 있는 것으로 가정
    # 과목을 가져오는 method
    @staticmethod
    def get_course():
        return "GEV6145"


    # 평균성적 출력하는method
    @staticmethod
    def get_average_score(stu_list):
        score_sum = 0
        for stu in stu_list:
            score_sum += stu.score
        score_avg = score_sum / len(stu_list)
        return score_avg
```

```python
# 3. [10] Excel 화일로 제공되는 수강생 명단을 읽어드려, 수강생 instance를 생성합니다.
#     생성되는 instance 명은 stu01부터 stu36으로 합니다. 제공되는 excel 화일의 '번호' column의 값-2가 stuXX의 XX에 해당합니다.
#     예: 번호 값이 38이면 stu36이 됩니다.

# 4. [10] 수강생 별 점수는 80점이상 100점이하 사이의
#    random number로(소수점이하 1자리) 생성하되 정규분포하도록 할당합니다.



 # Excel 파일을 읽어와서 df 에 저장 -> excel 파일은 같은 경로에 있다고 가정 -> 아래 코드는 코랩에서 사용한 경로
df = pd.read_excel('/content/drive/My Drive/GEV6145_students.xlsx')

# 생성된 인스턴스 담을 리스트
stu_list = []

# 반복문을 통해 stu01부터 stu36까지의 인스턴스를 생성 -> 번호의 -2가 stuXX의 XX에 해당
for i in range(len(df)):                                  # len(df) -> 총 36개
    number = int(df.iloc[i,0])                          # 번호 가져오기

    # 10보다 작을경우에는 0n의 형식으로 만들어줘야함
    if number < 10:
        number = f"0{number-2}"
    else:
        number = f"{number - 2}"
    student_var_name = f"stu{number}"   # 변수명 생성

    ID = df.iloc[i, 2]
    name = df.iloc[i, 3]
    score = round(random.uniform(80, 100), 1) # 랜덤함수로 80~100 점수 생성
    major = df.iloc[i, 1]
    course = Student.get_course()

    # 생성된 변수명으로 Studen instance 생성
    globals()[student_var_name] = Student(ID, name, score, major, course)

    # 생성된 instance 리스트에 넣기
    stu_list.append(globals()[student_var_name])

```

```python
# 5. [10] 생성된 총 수강생 수를 화면에 출력하십시오.
print(f'생성된 총 수강생 수: {len(stu_list)}')
```

```python
# 6. [10] 수강생들의 점수 평균을 화면에 출력하십시오. 80-100점 사이에서 정규분포하므로
#         평균은 90점에 가까울 것입니다.
average_score = Student.get_average_score(stu_list)
print(f"수강생들의 점수 평균: {average_score}점")
```
